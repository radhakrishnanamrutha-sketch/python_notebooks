```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
```

```python
def run_counts(qc,shots=1000):
  simulator = AerSimulator()
  job = simulator.run(qc, shots=shots)
  result = job.result()
  counts = result.get_counts()
  return counts
```

```python
#Exercise 1: Create Your First Quantum Circuit
from qiskit import QuantumCircuit
qc= QuantumCircuit(1)
print (qc)
qc.draw ('mpl', initial_state= True)
```

```python
#Exercise 2: Initialize a qubit in |0>
qc= QuantumCircuit(1)
print ("The qubit is automatically initialised to |0>")
qc.draw ('mpl', initial_state= True)
```

```python
#Exercise 3: Measure |0>
qc= QuantumCircuit(1,1)
qc.measure (0,0)
counts = run_counts (qc, shots=1000)
print (counts)
qc.draw ('mpl')
```

```python
#Add a second qubit
qc= QuantumCircuit(2,2)
qc.measure (0,0)
qc.measure (1,1)
counts = run_counts (qc, shots=1000)
print (counts)
qc.draw ('mpl')
```

```python
#Initialize |1> using X gate
qc = QuantumCircuit(1,1)
qc.x (0)
qc.measure (0,0)
for shots in (10,100,1000):
  counts = run_counts (qc, shots=shots)
  print (f"shots - {shots}, counts {counts}")
qc.draw ('mpl')
```

```python
#Measure |1>
qc= QuantumCircuit(1,1)
qc.x (0)
qc.measure (0,0)
counts = run_counts (qc, shots=1000)
print (counts)
qc.draw ('mpl')
```

```python
#Apply Hadamard Gate to 0>- Exercise 6
# create an equal superposition using Hadamard Gate
qc= QuantumCircuit(1,1)
qc.h (0)
qc.measure (0,0)
counts = run_counts (qc, shots=1000)
print (counts)
qc.draw ('mpl')
```

```python
# Apply Hadamard Gate to 1>- Exercise 7
# create an equal superposition using Hadamard Gate
qc = QuantumCircuit(1,1)
qc.x (0)
qc.h (0)
qc.measure (0,0)
counts = run_counts (qc, shots=1000)
print (counts)
qc.draw ('mpl')
```

```python
# Verify 50-50 Probability (Exercise 8 )
# Convert count to Probabilities
qc = QuantumCircuit(1,1)
qc.h (0)
qc.measure (0,0)
shots = 1000
counts = run_counts (qc, shots=1000)
print ("counts:",counts)
p0 = counts.get ('0',0)
p1 = counts.get ('1',1)
print("Probability of 0:",p0)
print("Probability of 1:", p1)
```

```python
#Compare 100, 1000, and 10000 shots-Exercise 9
qc = QuantumCircuit(1,1)
qc.h (0)
qc.measure (0,0)

for shots in (100,1000,10000):
  counts = run_counts (qc, shots=shots)
  p0 = counts.get ('0',0)
  p1 = counts.get ('1',1)
  print ("Probability of 0:",p0 )
  print ("Probability of 1:",p1)
  print ("Shots:", shots)
  print ("Count:", counts)
  print ("P(0)",p0)
  print ("P(1)",p1)
  print("-" * 30)
```

```python
# Statistical analysis of Hadamard measurement- Exercise 10
qc = QuantumCircuit(1,1)
qc.h (0)
qc.measure (0,0)
ideal = 0.5
for shots in [100,1000,10000]:
  counts = run_counts (qc, shots=shots)
  p0 = counts.get ('0',0)
  p1 = counts.get ('1',1)
  error_p0 = abs (p0 - ideal)
  error_p1 = abs (p1 - ideal)
  print (f"Shots: {shots}, P(0): {p0}, Error: {error_p0}")
  print (f"Shots: {shots}, P(1): {p1}, Error: {error_p1}")
```
